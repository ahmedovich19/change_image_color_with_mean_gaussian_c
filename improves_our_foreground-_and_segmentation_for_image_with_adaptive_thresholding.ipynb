```python
import cv2
from cv2 import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/sudoku.jpg',0)
_,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,111,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,111,2)
```

```python
cv2.imshow('image',img)
cv2.imshow('THRESH_BINARY',th1)
cv2.imshow('ADAPTIVE_THRESH_MEAN_C',th2)
cv2.imshow('ADAPTIVE_THRESH_GAUSSIAN_C',th3)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
#the main image colored with gray
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
#the image after implementing THRESH BINARY 
plt.imshow(cv2.cvtColor(th1,cv2.COLOR_BGR2RGB))
```

```python
#the image after implementing ADAPTIVE_THRESH_MEAN_C
plt.imshow(cv2.cvtColor(th2,cv2.COLOR_BGR2RGB))
```

```python
#the image after implementing ADAPTIVE_THRESH_GAUSSIAN_C
plt.imshow(cv2.cvtColor(th3,cv2.COLOR_BGR2RGB))
```
